```python
# what is series and how to create 
# series is one dimentional array
name = ["sancho", "white", "rice", "stone", "omahan", " rashford"]
s1= pd.Series(name)
s1
```

```python
# series as a tup
tup1 = ("one","two","three", "four")
s2= pd.Series(tup1)
s2
```

```python
# series as dictionary
dic =  {"name":"Sancho","age":12,"salary":324564}
pd.Series(dic)
```

```python
# series as dictionary with index
name = ["sancho", "white", "rice", "stone", "omahan", " rashford"]
s1= pd.Series(name, index= ["Y","W","R","T","C", "G"])
s1
```

```python
# creating dataframe from a list ( it is made up of multiple columns)

s1.to_frame(name="people's name")
            
```

```python
# creating dataframe from list with customized index
name = ["sancho", "white", "rice", "stone", "omahan", "rashford"]

pd.DataFrame(name, columns=["name"], index=["A","B","C","D","E","F"])
```

```python
# creating dataframe  from dictinary
dic =  {"name":["Sancho","white","rice","stone","onana","rashford"],
         "age":[12,22,34,25,26,27],
        "salary":[2345,3456,4567,3458,5678,7865]}
df=pd.DataFrame(dic)
```
